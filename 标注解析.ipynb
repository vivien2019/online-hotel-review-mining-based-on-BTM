```python
%matplotlib inline
import pandas as pd
```

```python
df=pd.read_excel('中文关键词标注.xlsx')
df
```

```python
d={}
for i in range(len(df)):
    row=df.iloc[i]
    if row['类别']!=0:
        d[row['keywords']]=row['类别']
```

```python
len(d)
```

```python
with open('中文主题词.txt','r',encoding='utf-8')as f:
    rows=f.readlines()
    dl=[]
    for i in range(len(rows)):
        row=rows[i]
        dd={}
        for r in row.split('\t')[1].split():
            k,v=r.split(':')
            if k in d:
                if d[k] in dd:
                    dd[d[k]]+=float(v)
                else:
                    dd[d[k]]=float(v)
        dl.append(dd)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x=MinMaxScaler().fit_transform(df_k)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabaz_score
```

```python
km = KMeans(n_clusters=5, random_state = 666)   #将数据集分为2类
y_pre = km.fit_predict(x)
print(calinski_harabaz_score(x, y_pre)) 
```

```python
for i in range(2,10):
    km = KMeans(n_clusters=i, random_state = 666)   #将数据集分为2类
    y_pre = km.fit_predict(x)
    print(i,calinski_harabaz_score(x, y_pre)) 
```

```python
df=pd.read_excel('英文关键词标注.xlsx')
df
```

```python
d={'experience':'卫生',
'facility':'设施',
'staff':'服务',
'price':'价位',
'internet':'服务',
'room':'卫生',
'food':'餐饮',
'environment':'环境',
'travel':'交通',
'location':'交通',
'service':'服务'}
```

```python
df=df[df['category']!=0]
```

```python
d
```

```python
df['category']=df['category'].apply(lambda x:x.strip())
```

```python
df['类别']=df['category'].apply(lambda x:d[x])
```

```python
df
```

```python
xs=[]
for k in  ['设施', '服务', '餐饮', '环境', '交通', '价位', '卫生']:
    x=df[df['类别']==k][['keywords','probability']].head(500)
    y=x.reset_index(drop=True)
    y.columns=[k+'_'+col for col in y.columns]
    xs.append(y)
ddf=pd.concat(xs,axis=1)
ddf
```
